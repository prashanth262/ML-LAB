## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'


df['Embarked'].fillna('S',inplace=True)
df.head()
```

```python
#Impute the missing values in Fare column with median value
df['Fare'].fillna(df['Fare'].median())
```

```python
#Impute the Age column. It has too many missing values but it may be important. 

print(df['Age'].fillna(df['Age'].mean()))



 
```

```python
#We fill the missing values with numbers randomly drawn from distribution
import numpy as np
age_count=df['Age'].isnull().sum()
random=np.random.choice(df['Age'].dropna(),size=age_count,replace=True)

df.loc[df['Age'].isnull(),'Age']=random

df.head()
```

```python
#that has same mean and standard deviation as non missing Age data
mean_age=df['Age'].mean()
std_age=df['Age'].std()

n_miss=df['Age'].isnull().sum()
random=np.random.normal(loc=mean_age,scale=std_age,size=n_miss)
random = np.clip(random, 0, 80)# limit of numbers


df.loc[df['Age'].isnull(),'Age']=random
df.head()
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
df['FamilySize']=df['SibSp']+df['Parch']+1
df.head()
#FamilySize = Parch + SibSp + 1
```

```python
# Create new feature IsAlone from FamilySize

#If FamilySize == 1 then IsAlone = True else IsAlone = False
df['IsAlone'] = df['FamilySize'].apply(lambda x: True if x == 1 else False)


df.head()
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin

df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'],inplace=True,errors='ignore')

df.head()
```

## Model Building

```python
#Prepare X->feature and y->traget
y=df['Survived']
X=df.drop(df['Survived'])
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder
cat=df.select_dtypes(include=['object']).columns
le=LabelEncoder()
for col in cat:
    df[col]=le.fit_transform(df[col])
df.head()
```

```python
# concat df_categorical with original df
import pandas as pd
cat=pd.get_dummies(df['Sex'],drop_first=True)
df=pd.concat([df,cat],axis=1)

df.head()
```

```python
# Drop all columns named exactly '1'
df = df.drop(columns=[col for col in df.columns if col == 1])

# Show the updated DataFrame
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create pairplot with KDE on the diagonals
sns.pairplot(X,diag_kind='kde')
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split
y=df['Survived']
X=df.drop('Survived',axis=1)
X_train,X_test,y_train,y_valid=train_test_split(X,y,test_size=0.2)
print(X_train.shape,X_train.shape,y_train.shape,y_valid.shape)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#Print the tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plot_tree(model,filled=True,feature_names=X.columns,class_names=['Not Survived','Survived'])
plt.show()

# Importing required packages for visualization
```

```python
# plotting tree with max_depth=3
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plot_tree(model,filled=True,feature_names=X.columns,class_names=['Not Survived','Survived'],max_depth=3)
plt.show()
```
